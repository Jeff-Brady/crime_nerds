```python
years_list = list(range(2010,2023))

years_list
```

```python

incomes = {}

for x in years_list:
    median_income = pd.read_csv(f'../GOODDATA/productDownload_2024-04-25T075900/ACSST5Y{x}.S1901-Data.csv')
    median_income = median_income[['NAME','S1901_C01_012E']]
    median_income['year'] = x
    median_income.rename(columns={'NAME':'borough','S1901_C01_012E':'median_income_usd'}, inplace=True)
    median_income.drop(median_income.index[0], inplace=True)

    incomes[f'median_income_{x}'] = median_income
```

```python
incomes['median_income_2014']
```

```python
new_incomes = pd.concat(incomes.values(), ignore_index=True)

new_incomes
```
